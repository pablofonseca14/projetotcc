```python
def soma_de_gols(campeonato):
  campeonato['Gols Casa'] = campeonato.groupby(['Casa'])['Placar Casa'].cumsum() - campeonato['Placar Casa']
  campeonato['Gols Visitantes'] = campeonato.groupby(['Visitante'])['Placar Visitante'].cumsum() - campeonato['Placar Visitante']
  campeonato['Gols Conc. Casa'] = campeonato.groupby(['Casa'])['Placar Visitante'].cumsum() - campeonato['Placar Visitante']
  campeonato['Gols Conc. Visitante'] = campeonato.groupby(['Visitante'])['Placar Casa'].cumsum() - campeonato['Placar Casa']

soma_de_gols(eng_18_19)
soma_de_gols(eng_17_18)
soma_de_gols(eng_16_17)
soma_de_gols(eng_15_16)
soma_de_gols(eng_14_15)
soma_de_gols(eng_13_14)
soma_de_gols(eng_12_13)
soma_de_gols(eng_11_12)
soma_de_gols(eng_10_11)
soma_de_gols(eng_09_10)
soma_de_gols(eng_08_09)
soma_de_gols(eng_07_08)
soma_de_gols(eng_06_07)
soma_de_gols(eng_05_06)
soma_de_gols(ita_18_19)
soma_de_gols(ita_17_18)
soma_de_gols(ita_16_17)
soma_de_gols(ita_15_16)
soma_de_gols(ita_14_15)
soma_de_gols(ita_13_14)
soma_de_gols(ita_12_13)
soma_de_gols(ita_11_12)
soma_de_gols(ita_10_11)
soma_de_gols(ita_09_10)
soma_de_gols(ita_08_09)
soma_de_gols(ita_07_08)
soma_de_gols(ita_06_07)
soma_de_gols(ita_05_06)
```

```python
def gols_marcados(campeonato):
    # Cria√ß√£o de dicion√°rio. key = times
    times = {}
    for i in campeonato.groupby('Casa').mean().T.columns:
        times[i] = []
    
    # Valor correspondente para cada chave. Cada time ter√° um GMC ou GMV
    for i in range(380):
        GMC = campeonato.iloc[i]['Placar Casa'] #Gols Marcados Casa
        GMV = campeonato.iloc[i]['Placar Visitante'] #Gols Marcados Visitante
        times[campeonato.iloc[i].Casa].append(GMC)
        times[campeonato.iloc[i].Visitante].append(GMV)
    
    # Cria√ß√£o de DF; linhas = times, colunas = rodada
    GolsMarcados = pd.DataFrame(data=times, index = [i for i in range(1,39)]).T
    GolsMarcados[0] = 0
    # Somar quantidade de gols da rodada atual com o somat√≥rio anterior
    for i in range(2,39):
        GolsMarcados[i] = GolsMarcados[i] + GolsMarcados[i-1]
    return GolsMarcados



# Mesmo procedimento para os gols concedidos
def gols_concedidos(campeonato):
    
    times = {}
    for i in campeonato.groupby('Casa').mean().T.columns:
        times[i] = []
    
    
    for i in range(380):
        GCC = campeonato.iloc[i]['Placar Visitante'] #Gols Concedidos Casa
        GCV = campeonato.iloc[i]['Placar Casa'] #Gols Concedidos Visitante
        times[campeonato.iloc[i].Casa].append(GCC)
        times[campeonato.iloc[i].Visitante].append(GCV)
    
    
    GolsConced = pd.DataFrame(data=times, index = [i for i in range(1,39)]).T
    GolsConced[0] = 0
     
    for i in range(2,39):
        GolsConced[i] = GolsConced[i] + GolsConced[i-1]
    return GolsConced


#cria√ß√£o de novas colunas com os novos dados
def gols_c_v(campeonato):
    GC = gols_concedidos(campeonato)
    GS = gols_marcados(campeonato)
   
    j = 0
    GMC = []
    GMV = []
    GCC = []
    GCV = []

    for i in range(380):
        ht = campeonato.iloc[i].Casa
        at = campeonato.iloc[i].Visitante
        GMC.append(GS.loc[ht][j])
        GMV.append(GS.loc[at][j])
        GCC.append(GC.loc[ht][j])
        GCV.append(GC.loc[at][j])
        
        if ((i + 1)% 10) == 0:
            j = j + 1
        
    campeonato['GMC'] = GMC
    campeonato['GMV'] = GMV
    campeonato['GCC'] = GCC
    campeonato['GCV'] = GCV
    
    return campeonato


# Aplica√ß√£o em cada dataset
gols_c_v(eng_18_19) 
gols_c_v(eng_17_18)
gols_c_v(eng_16_17)
gols_c_v(eng_15_16)
gols_c_v(eng_14_15)
gols_c_v(eng_13_14)
gols_c_v(eng_12_13)
gols_c_v(eng_11_12)
gols_c_v(eng_10_11)
gols_c_v(eng_09_10)
gols_c_v(eng_08_09)
gols_c_v(eng_07_08)
gols_c_v(eng_06_07)
gols_c_v(eng_05_06)
gols_c_v(ita_18_19)
gols_c_v(ita_17_18)
gols_c_v(ita_16_17)
gols_c_v(ita_15_16)
gols_c_v(ita_14_15)
gols_c_v(ita_13_14)
gols_c_v(ita_12_13)
gols_c_v(ita_11_12)
gols_c_v(ita_10_11)
gols_c_v(ita_09_10)
gols_c_v(ita_08_09)
gols_c_v(ita_07_08)
gols_c_v(ita_06_07)
gols_c_v(ita_05_06)
print('')


```

```python
#Adicionando coluna com a quantidade total de pontos que cada equipe marcou. At√© o momento temos quantos pontos cada equipe marcou no confronto atual,
#quantos pontos foram marcados de acordo com sua condi√ß√£o (Casa/Visitante)

def ponto_partida(resultado):
    if resultado == 'S':
        return 3
    elif resultado == 'E':
        return 1
    else:
        return 0
    

def total_pontos(equipe):
    pontos_equipe = equipe.applymap(ponto_partida)
    for i in range(2,39):
        pontos_equipe[i] = pontos_equipe[i] + pontos_equipe[i-1]
        
    pontos_equipe.insert(column =0, loc = 0, value = [0*i for i in range(20)])
    return pontos_equipe


def res_l(campeonato):
    
    time = {}
    for i in campeonato.groupby('Casa').mean().T.columns:
        time[i] = []



    
    for i in range(len(campeonato)):
        if campeonato.iloc[i].Resultado == 'C':
            time[campeonato.iloc[i].Casa].append('S')
            time[campeonato.iloc[i].Visitante].append('D')
        elif campeonato.iloc[i].Resultado == 'V':
            time[campeonato.iloc[i].Visitante].append('S')
            time[campeonato.iloc[i].Casa].append('D')
        else:
            time[campeonato.iloc[i].Visitante].append('E')
            time[campeonato.iloc[i].Casa].append('E')
            
    return pd.DataFrame(data=time, index = [i for i in range(1,39)]).T

def pts_c_v(campeonato):
    equipe = res_l(campeonato)
    soma_pontos = total_pontos(equipe)
    PC = [] #pontos casa
    PV = [] #pontos visitante
    j = 0
    for i in range(380):
        ptc = campeonato.iloc[i].Casa #pontos casa
        ptv = campeonato.iloc[i].Visitante #pontos visitante
        PC.append(soma_pontos.loc[ptc][j])
        PV.append(soma_pontos.loc[ptv][j])

        if ((i + 1)% 10) == 0:
            j = j + 1
            
    campeonato['Total P. Casa'] = PC
    campeonato['Total P. Visit.'] = PV
    return campeonato
  



pts_c_v(eng_18_19)
pts_c_v(eng_17_18)
pts_c_v(eng_16_17)
pts_c_v(eng_15_16)
pts_c_v(eng_14_15)
pts_c_v(eng_13_14)
pts_c_v(eng_12_13)
pts_c_v(eng_11_12)
pts_c_v(eng_10_11)
pts_c_v(eng_09_10)
pts_c_v(eng_08_09)
pts_c_v(eng_07_08)
pts_c_v(eng_06_07)
pts_c_v(eng_05_06)
pts_c_v(ita_18_19)
pts_c_v(ita_17_18)
pts_c_v(ita_16_17)
pts_c_v(ita_15_16)
pts_c_v(ita_14_15)
pts_c_v(ita_13_14)
pts_c_v(ita_12_13)
pts_c_v(ita_11_12)
pts_c_v(ita_10_11)
pts_c_v(ita_09_10)
pts_c_v(ita_08_09)
pts_c_v(ita_07_08)
pts_c_v(ita_06_07)
pts_c_v(ita_05_06)
print('')
```

```python
#Sequencia de resultados anteriores

def res_anteriores(campeonato,num):
    anterior = res_l(campeonato)
    res_anterior = anterior.copy()
    for i in range(num,39):
        res_anterior[i] = ''
        j = 0
        while j < num:
            res_anterior[i] += anterior[i-j]
            j += 1           
    return res_anterior

def ad_res(campeonato,num):
    anterior = res_anteriores(campeonato,num)
    h = ['-' for i in range(num * 10)]  # Resultado nulo, pois n√£o √© poss√≠vel determinar o √∫ltimo resultado dos 10 primeiros jogos, pen√∫ltimo resultado para os 20 primeiros jogos...
    a = ['-' for i in range(num * 10)]
    
    j = num
    for i in range((num*10),380):
        casa = campeonato.iloc[i].Casa
        visit = campeonato.iloc[i].Visitante
        
        ultimo = anterior.loc[casa][j]           
        h.append(ultimo[num-1])                
        
        ultimo = anterior.loc[visit][j]              
        a.append(ultimo[num-1])                   
        
        if ((i + 1)% 10) == 0:
            j = j + 1

    campeonato['C' + str(num)] = h                 
    campeonato['V' + str(num)] = a

    
    return campeonato


def ad_resultados_df(campeonatos):
    campeonatos = ad_res(campeonatos,1)
    campeonatos = ad_res(campeonatos,2)
    campeonatos = ad_res(campeonatos,3)
    return campeonatos   

ad_resultados_df(eng_18_19)
ad_resultados_df(eng_17_18)
ad_resultados_df(eng_16_17)
ad_resultados_df(eng_15_16)
ad_resultados_df(eng_14_15)
ad_resultados_df(eng_13_14)
ad_resultados_df(eng_12_13)
ad_resultados_df(eng_11_12)
ad_resultados_df(eng_10_11)
ad_resultados_df(eng_09_10)
ad_resultados_df(eng_08_09)
ad_resultados_df(eng_07_08)
ad_resultados_df(eng_06_07)
ad_resultados_df(eng_05_06)
ad_resultados_df(ita_18_19)
ad_resultados_df(ita_17_18)
ad_resultados_df(ita_16_17)
ad_resultados_df(ita_15_16)
ad_resultados_df(ita_14_15)
ad_resultados_df(ita_13_14)
ad_resultados_df(ita_12_13)
ad_resultados_df(ita_11_12)
ad_resultados_df(ita_10_11)
ad_resultados_df(ita_09_10)
ad_resultados_df(ita_08_09)
ad_resultados_df(ita_07_08)
ad_resultados_df(ita_06_07)
ad_resultados_df(ita_05_06)
print('')
```
